```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn import tree
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Load the diabetes Dataset set to try
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

#print(df.head(1))
```

```python
df.iloc[:10]
```

```python
# Arrange Data into Features Matrix and Target Vector
X = df.loc[:, df.columns != 'target']
y = df.loc[:, 'target'].values

# Split the data into training and testing sets for the classifier
X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state=0)

# Random Forests call
#rf = RandomForestClassifier(n_estimators=100, random_state=0)
ranforre = RandomForestClassifier(max_samples=0.2, max_features=5, n_estimators=25, random_state=0, max_depth=4, verbose=2)
ranforre.fit(X_train, Y_train)
```

```python
fn=data.feature_names
cn=data.target_names
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=800)
tree.plot_tree(ranforre.estimators_[24],
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('rf_individualtree.png')
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 5,figsize = (10,2), dpi=900)
for index in range(0, 5):
    tree.plot_tree(ranforre.estimators_[index],
                   feature_names = fn, 
                   class_names=cn,
                   filled = True,
                   ax = axes[index]);

    axes[index].set_title('rand_forrest: ' + str(index), fontsize = 11)
fig.savefig('ranforre_trees.png')
```
